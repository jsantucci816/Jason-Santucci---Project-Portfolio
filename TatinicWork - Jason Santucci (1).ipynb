```python
# Lets get the Data Ready
# Reformat the data into pandas dataframes to get ready for Machine Learning algorithms 
# We have to know the shape of these datasets Later when we join and split them

train_shape = train.shape 
train_rows = train.shape[0]
train_cols = train.shape[1]

print("The shape of train is " + str(train_shape))
print("Our training set has " + str(train_rows))
print ("Our training set has " + str(train_cols))
```

```python
# Do the same thing for the test rows

test_shape = test.shape 
test_rows = test.shape[0]
test_cols = test.shape[1]

print("The shape of test is " + str(test_shape))
print("Our testing set has " + str(test_rows))
print ("Our testing set has " + str(test_cols))
```

```python
combined = pd.concat((train,test)) # Combine the train and the test dataframes together
combined_shape = combined.shape
combined_rows = combined.shape[0]
combined_cols = combined.shape[1]
print("Our concatenated set has " + str(combined_rows) + " rows")
print("Our concatenated set has " + str(combined_cols) + " columns")
combined.head()
```

```python
# But what about the Survived column? It is in the train datset but not the test dataset.

combined.sample(100)
```
